```
import newspaper
from newspaper import Article
url='https://www.bbc.com/hindi/science-51366908'
a = Article(url, language='hi') # hindi
#for making object
a.download() #for downloading article
a.parse() #for parsing the article

print(a.text)
```

```
 import nltk
 nltk.download('punkt')
url='https://khabar.ndtv.com/news/india/ladakh-clash-army-chief-commends-soldiers-who-fought-in-face-offs-with-china-2251431'
toi_article = Article(url, language="hi") # en for English 
  
 
toi_article.download() #for downloading article
  
 
toi_article.parse() #for parsing the article
 
toi_article.nlp() 
  
 
print("Article's Title:") 
print(toi_article.title) 
 
  
 
print("Article's Text:") 
print(toi_article.text) #for extracting the text
 
 
print("Article's Summary:") 
print(toi_article.summary) #for getting summary of article
 
  
 
print("Article's Keywords:") 
print(toi_article.keywords) #article keywords sorted by frequency
```
